```python

import pandas as pd

# Tạo một dictionary với dữ liệu mẫu
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ivan', 'Jack', 'Kelly', 'Liam', 'Mona', 'Nina', 'Oscar'],
    'Age': [25,30,35,28,22,45,34,31,27,29,33,40,26,32,36],
    'Salary': [50000,60000,70000,55000,52000,80000,72000,68000,61000,59000,63000,77000,53000,66000,75000]
    
}
# Câu 1
df = pd.DataFrame(data)
```

```python
# Câu 7
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 6))
plt.bar(df['Name'], df['Age'], color='skyblue')
plt.title('Biểu đồ cột cho cột Age')
plt.show()
```

```python
# Câu 8
plt.figure(figsize=(15, 6))
plt.plot(df['Name'], df['Salary'],marker='s', linestyle='-', color='skyblue')
plt.title('Biểu đồ đường cho cột Salary')
plt.grid(True)
plt.show()
```

```python
# Câu 9
plt.figure(figsize=(15, 13))
plt.pie(df['Age'].value_counts(), labels= df['Age'].value_counts().index)
plt.title('Biểu đồ tròn cho cột Age')
plt.show()
```

```python
# Câu 10
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Salary'], color='skyblue')
plt.title('Biểu đồ phân tán giữa Age và Salary')
plt.grid(True)
plt.show()
```

```python
# câu 11
df.isna().any()
```

```python
# Câu 12
avg = df['Age'].mean()
df.loc[df['Age'] > 30, 'Age'] = avg
df.head(20)
```

```python
# Câu 13
min_age = df['Age'].min()
max_age = df['Age'].max()
df['Chuan_Hoa_Age'] = (df['Age'] - min_age) / (max_age - min_age)
df[['Name', 'Age', 'Chuan_Hoa_Age', 'Salary']].head
```

```python
# Câu 14
def Age(age):
    if age < 30:
        return 'Young'
    elif age >= 60:
        return 'Old'
    else:
        return 'Middle'

df['Age_group'] = df['Age'].apply(Age)
df.head
```

```python
# Câu 15
df['Salary_Pct_Change'] = df['Salary'].pct_change() * 100
df.head
```

```python
# Câu 16
df.drop_duplicates(subset='Name', keep='first').head
```

```python
# Câu 17
df.to_csv('baikiemtra.csv', index=False)
```
